````python
#Load in libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

## Bar chart using Hotdog Contest Winners dataset

````python
#Read in file and view example
hdwin = pd.read_excel("hotdog-contest-winners.xlsm")
hdwin.head()
````

````python
#Need to update two column names to make it easier to use
hdwin.rename(columns = {'Dogs eaten':'Dogs_eaten'}, inplace = True)
hdwin.rename(columns = {'New record':'New_record'}, inplace = True)
hdwin.head()
````

````python
#Bar chart
plt.bar(hdwin.Year, hdwin.Dogs_eaten, color = 'blue')
plt.xlabel("Year")
plt.ylabel("Number of dogs eaten by winner")
plt.title("Number of hotdogs eaten by the contest winner")
plt.show()
````

## Stacked bar chart using Fifa World Cups dataset

````python
#Reference dataset: https://www.kaggle.com/abecklas/fifa-world-cup?select=WorldCups.csv
#The hotdog places dataset didn't make sense so found a different one
#Read in data and get preview
wc = pd.read_csv("WorldCups.csv")
wc.head()
````

````python
#Stacked bar charts
plt.bar(wc.Winner, wc.GoalsScored, label = 'Goals Scored', color = 'green')
plt.bar( wc.Winner, wc.MatchesPlayed, label = '# Matches', color = 'red')
plt.xlabel("Winning Country")
plt.ylabel("Number of Goals Scored and Matches Played")
plt.legend(loc = 'upper left')
plt.xticks(rotation = 90, fontsize = 10)
plt.title('Winning Countries by # Goals Scored and # Matches Played')
plt.show()
````

## Pie chart using Obama Approval Ratings dataset

````python
#Read in dataset and get preview
oar = pd.read_excel("obama-approval-ratings.xls")
oar.head()
````

````python
#Create pie chart
#Referencing https://datatofish.com/pie-chart-matplotlib/
labels = oar.Issue
approve = oar.Approve
plt.pie(approve, labels = labels, autopct='%1.1f%%')
plt.title('Obamas Approval Rating by Issue')
plt.axis('equal')
plt.show()
````

## Donut chart using Obama Approval Ratings dataset

````python
#Donut graph
#Referencing: https://gist.github.com/krishnakummar/ad00d05311977732764f
labels = oar.Issue
sizes = oar.Disapprove
#Using same idea as pie chart
plt.pie(sizes, labels=labels,
        autopct='%1.1f%%', shadow=True)
        
#draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.title('Obama Disapproval Ratings by Issue')
plt.show()  
````
