````python
#Load in libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
#Read in file and make sure I did it right
wpop = pd.read_csv("world-population.csv")
wpop.head()
````

````python
#Line chart using Seaborn
sns.relplot(x='Year', y='Population',  estimator=None, kind='line', data=wpop)
````

````python
#Line chart using matplotlib
plt.plot(wpop['Year'], wpop['Population'], color = 'red')
plt.xlabel('Year')
plt.ylabel('Population in Billions')
plt.title('World Population by Year')
plt.grid(True)
plt.show()
````

````python
#Step chart using world population data. Doesn't look great so going to try with another dataset
x = wpop['Year']
y = wpop['Population']

plt.step(x,y)
plt.xlabel('Year')
plt.ylabel('Population in Billions')
plt.title('World Population by Year')
plt.grid(True)
plt.show()
````

````python
#Load in data
hdwin = pd.read_csv("hotdog-contest-winners.csv")
hdwin.head()
````

````python
#Step chart with hotdog data
x = hdwin['Year']
y = hdwin['Dogs eaten']

plt.step(x,y)
plt.xlabel('Year')
plt.ylabel('Number of Dogs Eaten')
plt.title('Number of Hotdogs Eaten by Year')
plt.show()
````
